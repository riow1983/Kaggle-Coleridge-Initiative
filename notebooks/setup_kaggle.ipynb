# Push Local Notebook to Kaggle

```python
%cd /content/drive/MyDrive/colab_notebooks/kaggle/Kaggle-Coleridge-Initiative/notebooks
```

```python
!ls
```

```python
notebook = "localnb003-spacy-dataset.ipynb"
kname = "l2knb003-spacy-dataset"
!mkdir {kname}
!cp {notebook} ./{kname}/{notebook} #バイナリファイルでコピーされる
!kaggle kernels init -p {kname}

with open(f"{kname}/kernel-metadata.json", "r") as jsonFile:
    data = json.load(jsonFile)

data["id"] = f"riow1983/{kname}"
data["title"] = kname
data["kernel_type"] = "notebook"
data["code_file"] = f"/content/drive/MyDrive/colab_notebooks/kaggle/Kaggle-Coleridge-Initiative/notebooks/{kname}/{notebook}"
data["language"] = "python"
data["competition_sources"] = ["coleridgeinitiative-show-us-the-data"]

with open(f"{kname}/kernel-metadata.json", "w") as jsonFile:
    json.dump(data, jsonFile)

!kaggle kernels push -p {kname}
```
